#Analyze Datas

```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)

# Paths to the datasets
weather_data_path = '/content/drive/MyDrive/TB3.0/weather_prediction_dataset.csv'
labels_data_path = '/content/drive/MyDrive/TB3.0/weather_prediction_bbq_labels.csv'

# Load the datasets
weather_data = pd.read_csv(weather_data_path)
labels_data = pd.read_csv(labels_data_path)

# Display dataset shapes
print("Weather Data Shape:", weather_data.shape)
print("Labels Data Shape:", labels_data.shape)

# Display column names
print("Weather Data Columns:", weather_data.columns.tolist())
print("Labels Data Columns:", labels_data.columns.tolist())

# Display first few rows of weather data
print("First few rows of Weather Data:")
print(weather_data.head())

# Display first few rows of labels data
print("First few rows of Labels Data:")
print(labels_data.head())

# Convert 'DATE' to datetime format
weather_data['DATE'] = pd.to_datetime(weather_data['DATE'], format='%Y%m%d')

# Function to plot weather data for a specific city
def plot_city_weather(city_name):
    # Check if columns exist before plotting
    temp_col = f'{city_name}_temp_mean'
    humidity_col = f'{city_name}_humidity'
    precipitation_col = f'{city_name}_precipitation'

    if temp_col in weather_data.columns:
        # Plot mean temperature over time
        plt.figure(figsize=(14, 8))
        plt.plot(weather_data['DATE'], weather_data[temp_col], label=f'{city_name} Mean Temperature', color='blue')
        plt.xlabel('Date', fontsize=14)
        plt.ylabel('Mean Temperature (°C)', fontsize=14)
        plt.title(f'Mean Temperature Over Time in {city_name}', fontsize=16)
        plt.legend()
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        print(f"Temperature data for {city_name} is not available.")

    if humidity_col in weather_data.columns:
        # Plot humidity over time
        plt.figure(figsize=(14, 8))
        plt.plot(weather_data['DATE'], weather_data[humidity_col], label=f'{city_name} Humidity', color='green')
        plt.xlabel('Date', fontsize=14)
        plt.ylabel('Humidity (%)', fontsize=14)
        plt.title(f'Humidity Over Time in {city_name}', fontsize=16)
        plt.legend()
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        print(f"Humidity data for {city_name} is not available.")

    if precipitation_col in weather_data.columns:
        # Plot precipitation over time
        plt.figure(figsize=(14, 8))
        plt.plot(weather_data['DATE'], weather_data[precipitation_col], label=f'{city_name} Precipitation', color='red')
        plt.xlabel('Date', fontsize=14)
        plt.ylabel('Precipitation (cm)', fontsize=14)
        plt.title(f'Precipitation Over Time in {city_name}', fontsize=16)
        plt.legend()
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        print(f"Precipitation data for {city_name} is not available.")

# List of cities based on columns
cities = [
    'BASEL', 'BUDAPEST', 'DE_BILT', 'DRESDEN', 'DUSSELDORF', 'HEATHROW', 'KASSEL',
    'LJUBLJANA', 'MAASTRICHT', 'MALMO', 'MONTELIMAR', 'MUENCHEN', 'OSLO',
    'PERPIGNAN', 'ROMA', 'SONNBLICK', 'STOCKHOLM', 'TOURS'
]

# Analyze weather data for all cities
for city in cities:
    plot_city_weather(city)
```

#Heathrow weather

```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose

# Step 1: Mount Google Drive
print("Step 1: Mounting Google Drive")
drive.mount('/content/drive', force_remount=True)

# Step 2: Load the datasets
print("Step 2: Loading the datasets")
weather_data_path = '/content/drive/MyDrive/TB3.0/weather_prediction_dataset.csv'
labels_data_path = '/content/drive/MyDrive/TB3.0/weather_prediction_bbq_labels.csv'

weather_data = pd.read_csv(weather_data_path)
labels_data = pd.read_csv(labels_data_path)

# Convert 'DATE' to datetime format
print("Converting 'DATE' to datetime format")
weather_data['DATE'] = pd.to_datetime(weather_data['DATE'], format='%Y%m%d')

# Step 3: Extract relevant columns for London (Heathrow)
print("Step 3: Extracting relevant columns for London (Heathrow)")
london_data = weather_data[['DATE', 'HEATHROW_temp_mean', 'HEATHROW_humidity', 'HEATHROW_precipitation']].dropna()

# Rename columns for convenience
london_data.columns = ['Date', 'Temperature', 'Humidity', 'Precipitation']

# Display the first few rows of the dataset
print("First few rows of London data:")
print(london_data.head())

# Step 4: Scatter Plot Matrix
print("Step 4: Creating Scatter Plot Matrix")
sns.pairplot(london_data, diag_kind='kde')
plt.suptitle("Scatter Plot Matrix for London Weather Data", y=1.02)
plt.show()

# Step 5: Correlation Matrix
print("Step 5: Creating Correlation Matrix")
correlation_matrix = london_data.corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the Correlation Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5, linecolor='white', annot_kws={"size": 14})
plt.title("Correlation Matrix for London Weather Data", fontsize=16)
plt.show()

# Step 6: Time Series Analysis
print("Step 6: Creating Time Series Analysis Plot")
plt.figure(figsize=(14, 8))
plt.plot(london_data['Date'], london_data['Temperature'], label='Temperature', color='blue', alpha=0.7)
plt.plot(london_data['Date'], london_data['Humidity'], label='Humidity', color='green', alpha=0.7)
plt.plot(london_data['Date'], london_data['Precipitation'], label='Precipitation', color='red', alpha=0.7)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.title('Time Series Analysis of Temperature, Humidity, and Precipitation in London', fontsize=16)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Step 7: Linear Regression Analysis
print("Step 7: Performing Linear Regression Analysis")

# Function to perform regression analysis and print summary
def perform_regression_analysis(independent_var, dependent_var):
    X = london_data[[independent_var]]
    Y = london_data[dependent_var]
    X = sm.add_constant(X)
    model = sm.OLS(Y, X).fit()
    print(f"\nRegression Analysis Summary for {independent_var} vs {dependent_var}:")
    print(model.summary())

# Temperature vs Humidity
perform_regression_analysis('Temperature', 'Humidity')
# Temperature vs Precipitation
perform_regression_analysis('Temperature', 'Precipitation')
# Humidity vs Precipitation
perform_regression_analysis('Humidity', 'Precipitation')

# Step 8: Seasonal Decomposition
print("Step 8: Performing Seasonal Decomposition")
london_data.set_index('Date', inplace=True)

def plot_seasonal_decompose(data, title):
    result = seasonal_decompose(data, model='additive', period=365)
    result.plot()
    plt.suptitle(title, y=1.02, fontsize=16)
    plt.show()

plot_seasonal_decompose(london_data['Temperature'], "Seasonal Decomposition of Temperature in London")
plot_seasonal_decompose(london_data['Humidity'], "Seasonal Decomposition of Humidity in London")
plot_seasonal_decompose(london_data['Precipitation'], "Seasonal Decomposition of Precipitation in London")

# Step 9: Lagged Correlations
print("Step 9: Calculating Lagged Correlations")
lags = range(1, 31)
temp_precip_corr = [london_data['Temperature'].shift(lag).corr(london_data['Precipitation']) for lag in lags]
humidity_precip_corr = [london_data['Humidity'].shift(lag).corr(london_data['Precipitation']) for lag in lags]
temp_humidity_corr = [london_data['Temperature'].shift(lag).corr(london_data['Humidity']) for lag in lags]

plt.figure(figsize=(14, 8))
plt.plot(lags, temp_precip_corr, label='Temperature vs Precipitation', color='blue', marker='o')
plt.plot(lags, humidity_precip_corr, label='Humidity vs Precipitation', color='green', marker='o')
plt.plot(lags, temp_humidity_corr, label='Temperature vs Humidity', color='orange', marker='o')
plt.xlabel('Lag (days)', fontsize=14)
plt.ylabel('Correlation', fontsize=14)
plt.title('Lagged Correlations of Temperature, Humidity, and Precipitation', fontsize=16)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
